```python


def bubble_sort(v):
  fim = len(v)
  while fim>0:
    i = 0
    while i < fim - 1:
      if v[i]  > v[i +1]:
        v[i], v[i + 1] = v[i + 1], v[i]
      i += 1
    fim -= 1
  return v

print(bubble_sort(random_numbers))




```
