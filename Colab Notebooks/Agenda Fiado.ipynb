```python
def salvar_em_arquivo(lista):
    arquivo = open("agenda.txt", "w")

    for cliente in lista:
        arquivo.write(
            "{},{}\n".format(cliente['Nome'], cliente['Saldo'],))

    arquivo.close()
    


# Função para carregar arquivo salvo na hora de inicializar o programa.
def carregar_clientes():
    lista = []

    try:
        arquivo = open("agenda.txt", "r")

        for linha in arquivo.readlines():
            coluna = linha.strip().split(",")

            cliente = {
                "Nome": coluna[0],
                "Saldo": coluna[1]
            }

            lista.append(cliente)

        arquivo.close()
    except FileNotFoundError:
        pass

    return lista


# Função para transformar o Nome do cliente em uma espécie de "id" único para cada cliente salvo.
def cliente_salvo(lista, nome):
    if len(lista) > 0:
        for cliente in lista:
            if cliente['Nome'] == nome:
                return True

    return False

#Função inserir novo cliente
def inserir(lista):
    while True:
        nome = input("Digite o nome do cliente: ")

        if not cliente_salvo(lista, nome):
            break

        else:
            print("Esse cliente já está salvo.")
            print("Verifique e tente novamente.")

    cliente = {"Nome": nome,
               "Saldo": input("Digite o saldo do cliente: ")
               }
    lista.append(cliente)
    print("O cliente {} foi salvo com sucesso! \n".format(cliente['Nome']))


#função alterar cliente existente
def alterar(lista):
    print(" Alterar cliente ")
    if len(lista) > 0:
        nome = input("Digite o nome do cliente: ")
        if cliente_salvo(lista, nome):
            print("O cliente foi encontrado. As informações seguem abaixo: ")
            for cliente in lista:
                if cliente['Nome'] == nome:
                    print("\tNome: {}".format(cliente['Nome']))
                    print("\tSaldo: {}".format(cliente['Saldo']))                    
                    print("=-=" * 12)
                    
                    cliente['Saldo'] = input("Digite o novo saldo do cliente: ")

                    print("Os dados do cliente {}, foram alterados com sucesso!".format(cliente['Nome']))
                    break
        else:
            print("Não há nenhum cliente com o nome {}.\n".format(nome))

    else:
        print("Ainda não há clientes. \n")

#Função limpar cliente
def remover(lista):
    print(" Remover cliente ")
    if len(lista) > 0:
        nome = input("Digite o nome do cliente: ")
        if cliente_salvo(lista, nome):
            print("O cliente foi encontrado. As informações seguem abaixo: ")
            for i, cliente in enumerate(lista):
                if cliente['Nome'] == nome:
                    print("\tNome: {}".format(cliente['Nome']))
                    print("\tSaldo: {}".format(cliente['Saldo']))
                    print("=-=" * 12)

                    del lista[i]

                    print("cliente removido")
                    break
        else:
            print("Não há nenhum cliente com o nome {}.\n".format(nome))

    else:
        print("Ainda não há clientes. \n")

#Função consultar cliente
def consultar(lista):
    print(" Consultar cliente ")
    if len(lista) > 0:
        nome = input("Digite o nome do cliente: ")
        if cliente_salvo(lista, nome):
            print("O cliente foi encontrado. As informações seguem abaixo: ")
            for cliente in lista:
                if cliente['Nome'] == nome:
                    print("Nome: {}".format(cliente['Nome']))
                    print("Saldo: {}".format(cliente['Saldo']))
                    print("=-=" * 12)
                    break
        else:
            print("Não há nenhum cliente com o nome {}.\n".format(nome))

    else:
        print("Ainda não há clientes. \n")

#Função relatório
def relatorio(lista):
    print(" clientes ")
    if len(lista) > 0:
        for i, cliente in enumerate(lista):
            print("cliente {}:".format(i + 1))
            print("\tNome: {}".format(cliente['Nome']))
            print("\tSaldo: {}".format(cliente['Saldo']))
            print("=-=" * 12)
        print("Quantidade de clientes: {}".format(len(lista)))
    else:
        print("Ainda não há clientes. \n")


#Função menu de opções
def menu():
    lista = carregar_clientes()

    while True:
        print("Agenda Fiado")
        print("=-=" * 12)
        print('''    [1] - Inserir cliente
    [2] - Consultar cliente
    [3] - Remover cliente
    [4] - Alterar cliente
    [5] - Relatório
    [6] - Sair''')
        opção = int(input("Digite o número da opção: "))

        if opção == 1:
            inserir(lista)
            salvar_em_arquivo(lista)

        elif opção == 2:
            consultar(lista)
        elif opção == 3:
            remover(lista)
            salvar_em_arquivo(lista)
        elif opção == 4:
            alterar(lista)
            salvar_em_arquivo(lista)
        elif opção == 5:
            relatorio(lista)
        elif opção == 6:
            print("Saindo da agenda...")
            break
        else:
            print("Opção inválida. Tente novamente: ")


menu()
```
